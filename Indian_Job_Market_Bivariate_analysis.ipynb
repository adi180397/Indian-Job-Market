```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/india_job_market_dataset.csv')
```

# **Bivariate Analysis**

```python
df.columns
```

```python
cross_tab = pd.crosstab(df['Job Title'], df['Job Location'])

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, cmap="YlGnBu", annot=False)
plt.title("Job Title vs Job Location")
plt.xlabel("Job Location")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

This heatmap shows which job titles are most popular in different locations.

```python
df['Salary Range'].value_counts()
```

```python
df['min']=df['Salary Range'].str.replace('LPA','').str.replace('+','').str.split('-',expand=True).astype(float)[0]
```

```python
df['max']=df['Salary Range'].str.replace('LPA','').str.replace('+','').str.split('-',expand=True).astype(float)[1]
```

```python
df.head()
```

```python
df['Average Salary']=df[['min','max']].mean(axis=1)
```

```python
df.sample(5)
```

```python
df[df['Job Title']=='Data Scientist']
```

```python
df[df['Job Title']=='Data Scientist']['Average Salary'].min()
```

```python
# Check if the data is correctly grouped by Job Title
grouped_data = df.groupby('Job Title')['Average Salary'].describe()
print(grouped_data)
```

```python
df[['Job Title','Average Salary']]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Job Title', y='Average Salary', palette='viridis')
plt.title('Boxplot of Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (in LPA)')
plt.xticks(rotation=45)
plt.show()
```

After doing analysis, it is concluded that each job title has an average salary 11 LPA(approx) and median salary 10 LPA

```python
plt.figure(figsize=(8, 6))
df.groupby('Remote/Onsite')['Number of Applicants'].mean().plot(kind='bar', color='skyblue')
plt.title("Average Number of Applicants by Remote/Onsite")
plt.ylabel("Average Number of Applicants")
plt.show()
```

Number of applicants are more for hybrid followed by remote jobs

```python
df['min exp req']=df['Experience Required'].str.replace('years','').str.replace('+','').str.split('-',expand=True).astype('float')[0]
```

```python
df['desirable exp req']=df['Experience Required'].str.replace('years','').str.replace('+','').str.split('-',expand=True).astype('float')[1]
```

```python
df['Average experience req']=df[['min exp req','desirable exp req']].mean(axis=1)
```

### Sactter plot between minimum experience required vs average salary

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='min exp req', y='Average Salary', hue='Job Type')
plt.title("Salary Range vs Experience Required")
plt.show()
```

It is not very clear that with the increase in experience the salary also increases
